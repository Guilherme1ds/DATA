```python
# Para rodar a aplicaÃ§Ã£o
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

# ConfiguraÃ§Ã£o visual
sns.set_theme(style="darkgrid")

# Carregar base de dados
# Como o arquivo estÃ¡ na mesma pasta, basta colocar o nome
df = pd.read_csv('titanic.csv')

# Mostra as primeiras 5 linhas formatadas
display(df.head())
# DICA: No VS Code, use 'display()' ao invÃ©s de 'print()' para tabelas bonitas

# Verificando tipos de dados e nulos
df.info()

# AnÃ¡lise da Idade
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', bins=30, kde=True, color='skyblue')
plt.title('DistribuiÃ§Ã£o de Idade dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Contagem')
plt.show()

# ComparaÃ§Ã£o de Tarifa por Classe
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Pclass', y='Fare', palette='Set2')
plt.title('PreÃ§o da Passagem por Classe (Bilhetes)')
plt.show()

# Apenas a linha de densidade (sem barras)
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='Age', hue='2urvived', fill=True, common_norm=False, palette='crest')
plt.title('Densidade de Idade: Quem Sobreviveu (1) vs Morreu (0)')
plt.show()

# Histograma de Idade separado por Sexo
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='Sex', multiple='stack', bins=25, palette='pastel')
plt.title('DistribuiÃ§Ã£o de Idade: Homens vs Mulheres')
plt.xlabel('Idade')
plt.show()

# PreÃ§o x Classe, separado por Sobrevivencia
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Pclass', y='Fare', hue='2urvived', palette='Set3')
plt.title('Impacto do PreÃ§o na SobrevivÃªncia por Classe')
plt.xlabel('Classe')
plt.ylabel('PreÃ§o da Passagem')
plt.legend(title='Sobreviveu?')
plt.show()

# Violin Plot: A evoluÃ§Ã£o do Box-plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Pclass', y='Age', split=True, hue='Sex', palette='muted')
plt.title('DistribuiÃ§Ã£o de Idade por Classe e Sexo')
plt.show()
```
