```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

sns.set_theme(style="darkgrid")

df = pd.read_csv('superstore.csv')

display(df.head())

df.info()

# Anlise de Vendas
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Sales', bins=30, kde=True, color='skyblue')
plt.title('Distribuio de Vendas')
plt.xlabel('Valor')
plt.ylabel('Contagem')
plt.show()

# Comparao de Vendas por Regio
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Region', y='Sales', palette='Set2')
plt.title('Vendas por Regio')
plt.show()

# Apenas a linha de densidade (sem barras)
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='Sales', hue='Category', fill=True, common_norm=False, palette='crest')
plt.title('Densidade de Vendas separada por Categoria de Produto')
plt.show()

# Histograma de Quantidade separado por Modo de Envio
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Quantity', hue='Ship Mode', multiple='stack', bins=10, palette='pastel')
plt.title('Distribuio de Quantidade separada por Modo de Envio')
plt.xlabel('Quantidade')
plt.show()

# Box Plot: Lucro por Regio, separado por Categoria
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Region', y='Profit', hue='Category', palette='Set3')
plt.title('Distribuio de Lucro por Regio e Categoria')
plt.xlabel('Regio')
plt.ylabel('Lucro')
plt.legend(title='Categoria')
plt.ylim(-1500, 1500)
plt.show()

# Violin Plot: A evoluo do Box-plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Sales', y='Category', split=True, hue='Sub-Category', palette='muted')
plt.title('Distribuio de Vendas por Categoria e Sub-Categoria')
plt.show()
```
